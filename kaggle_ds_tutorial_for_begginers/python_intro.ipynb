## Matplotlib
# Is a library that help us to plot the data

The easiest and most basic plots are line, scatter and histogram plots.

Line plot is better when x axis is time.
Scatter is better when there is correlation between two variables
Histogram is better when we need to see distribution of numerical data.
Customization: Colors,labels,thickness of line, title, opacity, grid, figsize, ticks of axis and linestyle

```python
# Line Plot Defede x Speed data columns
# color = color, label = label, linewidth = width of line, alpha = opacity, grid = grid, linestyle = sytle of line

data.Speed.plot(kind="line", color="g", label='Speed', linewidth=1, alpha=0.5, grid=True, linestyle=":")
data.Defense.plot(color="r", label="Defense", linewidth=1, alpha=0.5, grid=True, linestyle= "-.")

# Add legend into the plot
# Legend location possibilities:
# 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
plt.legend(loc="best")


plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Lini Line Plot - Speed x Defense")
plt.show()
```

```python
# Scatter Plot 
# Scatter is better when there is correlation between two variables

data.plot(kind="scatter", x="Attack", y="Defense", alpha=0.5, color="red")
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.title("Attack Defense Scatter Plot")
```

```python
# Histogram 
# Histogram is better when we need to see distribution of numerical data.
# bins = number of bar in figure

data.Speed.plot(kind="hist", bins=50, figsize=(12,20))
plt.show()
```

```python
# clf() = cleans it up again you can start a fresh
data.Speed.plot(kind = 'hist',bins = 50)
# CLF -> Clear  current figure
plt.clf()
```

# Pandas

```python
# Series type is a onde-dimentional array of data
series = data['Defense'] 
print(type(series))

# Dataframe is a 2-dimentional labeled data structure with columns with pottentially different types
data_frame = data[['Defense']]  # data[['Defense']]
print(type(data_frame))

# If we access the dataframe directly, this will return a array of its content
#Acessing like  data[['Defense']] , it will retun a dataframe
```

```python
# 1 - Filtering Pandas data frame
# There are only 3 pokemons who have higher defense value than 200
#This code filter the pokemons
logic_response = data['Defense']>200    
# logic_response is a boolean array with True and False register for each row, checking if the row  condition is true
print(logic_response)
print(type(logic_response))
data[logic_response]
```

```python
# 2 - Filtering pandas with logical_and
# There are only 2 pokemons who have higher defence value than 200 and higher attack value than 100

logic_response = np.logical_and(data['Defense']>200, data['Attack']>100)
print(logic_response)
print(type(logic_response))
data[logic_response]
```

```python
# This is also same with previous code line. Therefore we can also use '&' for filtering.
data[(data['Defense']>200) & (data['Attack']>100)]
```

```python
# Lambda Function 

square = lambda x: x**2     # where x is name of argument
print(square(4))
tot = lambda x,y,z: x+y+z   # where x,y,z are names of arguments
print(tot(1,2,3))

```

```python
# Conditionals on iterable
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```
